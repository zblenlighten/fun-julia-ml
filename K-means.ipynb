K Means, K = 3

```julia
data = copy(convert(Array{Float64}, df)')
```

```julia
seeds = initseeds(:rand, data, 3)
```

```julia
data_kmeans = kmeans(data, 3)
```

```julia
nclusters(data_kmeans)
```

```julia
counts(data_kmeans)
```

```julia
df.group = assignments(data_kmeans)
```
